### 서비스 가격 출력 프로그램 완성하기

```python
def service_price():
    service = input("서비스 종류를 입력하세요, a/b/c: ")
    valueAdded = input('부가세를 포함합니까? y/n : ')
    if valueAdded == 'y':
        if service == 'a':
            result = 23 * 1.1
        if service == 'b':
            result = 40 * 1.1
        if service == 'c':
            result = 67 * 1.1
    if valueAdded == 'n':
        if service == 'a':
            result = 23
        if service == 'b':
            result = 40
        if service == 'c':
            result = 67
    print(round(result, 1), "만 원입니다.")
```

```python
service_price()
```

```python
service_price()
```

## return 완벽 이해하기

```python
#print하는 함수
def addPrint(a,b):
    print(a+b)
```

```python
#return하는 함수
def addReturn(a,b):
    return a+b
```

```python
addPrint(1,2)
addReturn(1,2)
#별 차이 없음
```

```python
print("The result is", addPrint(1,2))
#print는 결괏값을 출력할 뿐 저장하지는 않는다.
```

```python
print("The result is", addReturn(1,2))
#return은 결괏값을 함수의 결과물로 '반환'하기 때문에 결괏값을 다른 곳에 활용가능
#즉 반환 = 최종 결과물을 도출하여 저장한다.
```

## 정규표현식으로 문자열 다루기

```python
import re
```

### match 메서드 - 문자열 도입에서 패턴찾기

##### match 메서드 사용법 : re.match(패턴, 문자열)

```python
re.match(r'life','life').group()
#문자열 'life'에서 'life'라는 패턴을 찾아 반환하세요
```

```python
def refinder(pattern, script):     #refinder라는 함수를 정의합니다.
    if re.match(pattern, script):  #원고(script)에서 패턴을 찾을 수 있다면
        print("Match!")            #'Match!'를 출력하세요
    else:
        print("Not a match!")      #패턴을 찾지 못했다면 'Not a match'를 출력하세요
```

```python
pattern = r'Life'
script = 'Life is so cool'
```

```python
refinder(pattern,script)
```

```python
pattern = r'life'
script = 'Life is so cool'
```

```python
refinder(pattern,script)
#match는 대소문자를 구분한다.
```

```python
pattern = r'is'
script = 'Life is so cool'
```

```python
refinder(pattern,script)
#match는 문자열의 시작부터 매칭인지 찾기때문에 중간에 있는 패턴을 찾지 못함
```

### search 메서드 - 문자열 전체에서 패턴찾기

##### search 메서드 사용법 : re.search(패턴, 문자열)

```python
pattern = r'is'
script = 'Life is so cool'
```

```python
re.search(pattern,script).group()    #script에서 패턴('r.is')을 찾아 반환하세요
#match는 중간에 있는 is를 찾지 못했으나 search는 가능!
```

### findall 메서드 - 패턴을 모두 찾아 리스트로 반환하기

##### findall 메서드 사용법 : re.findall(패턴, 찾으려는 문자열)

```python
number = 'My number is 511223-1****** and yours is 521012-2******'
re.findall('\d{6}',number)     #'\d{6}'패턴은 숫자를 여섯번 반복한 값을 의미합니다.
#match, search와 달리 findall은 결괏값이 빈 리스트로 나온다
```

```python
#정규표현식의 탐욕 = 모든 문자를 다 집어삼키는 것
```

```python
example1 = '저는 91년에 태어났습니다. 97년에는 IMF가 있었습니다. 지금은 2022년입니다.'
re.findall(r'\d.+년',example1)
#숫자(\d)로 시작하고, 어떤 문자든(.) 반복(+)되며, '년'으로 끝나는 문자열을 반환하라고 명령
#년으로 시작하는 문장이 통채로 출력되는 문제(탐욕)발생
```

```python
#탐욕을 멈추는 방법 = ?(물음표) 집어넣기
```

```python
re.findall(r'\d.+?년',example1)
```

```python
example = '이동민 교수는 다음과 같이 설명했습니다.(이동민, 2019). 그런데 다른 학자는 이 문제에 대해서 다른 견해를 가지고 있었습니다.(최재영, 2019). 또 다른 견해도 있었습니다.(Lion, 2018).'
```

```python
result = re.findall(r'\(.+\)',example)
```

```python
result
```

```python
#?(물음표)를 통해 탐욕 없애기
```

```python
result = re.findall(r'\(.+?\)',example)
#닫는 괄호 앞에 ?를 달아줌
```

```python
result
```

### split 메서드 - 문장 나누는 패턴만들기

#### split 메서드 사용법 : re.split(패턴, 문자열)

```python
sentence = 'I have a lovely dog, really. I am not telling a lie. What a pretty dog! I love this dog.'
```

```python
re.split(r'[.?!]', sentence)
#split은 특정한 패턴이 등장할 때 문자열을 나눈다.
#r을 쓰면 마침표, 느낌표, 물음표마다 역슬래시(\)를 붙이지 않아도 된다.
```

```python
data = 'a:3; b:4; c:5'
for i in re.split(r';', data):
    print(re.split(r':',i))
```

### sub 메서드 - 문자열 바꾸기

#### sub 메서드 사용법 : re.sub(찾을 패턴, 대체할 문자, 찾을 문자열)

```python
sentence = 'I have a lovely dog, really. I am not telling a lie. What a pretty dog! I love this dog.'
#이 문장에서 개(dog)를 고양이(cat)으로 고쳐보자
```

```python
re.sub(r'dog', 'cat', sentence)
```

```python
#불필요한 공백을 없애는 데에 아주 유용하게 사용될 수 있음
```

```python
words = 'I am home now. \n\n\nI am with my cat.\n\n'
print(words)
```

```python
re.sub(r'\n','',words)
```
