### 2)등장인물 리스트 만들기

```python
#대본에서 우린 패턴을 찾을 수 있다
#Monica:와 같이 대문자로 시작+소문자 반복+콜론(:)
```

```python
#등장인물 이름 모으기
char = re.compile(r'[A-Z][a-z]+:')
#[A-Z]는 대문자 전체, [a-z]는 소문자 전체를 의미
```

```python
#findall을 사용해 '등장인물:'형태로 된 패턴문자열 출력하기
re.findall(char, script101)
```

```python
#set키워드로 중복된 원소 지우기
a = [1,2,3,4,5,2,2]
set(a)
```

```python
set(re.findall(char, script101))
```

```python
#슬라이싱으로 특정위치의 문자 지우기
y = set(re.findall(char, script101))      #등장인물 리스트를 집합으로 바꿔 y 객체에 저장
z = list(y)                               #for문에 사용하기 위해 다시 리스트로 바꾸기
character = []                            #등장인물 리스트를 새로 담을 객체 character만들기
for i in z:
    character += [i[:-1]]                 #맨 뒤 콜론(:)을 지우고 charactor에 저장
```

```python
character
```

```python
#모든 명령을 한 줄로 나타내기
character = [x[:-1] for x in list(set(re.findall(r'[A-Z][a-z]+:',script101)))]
character
```

```python
#지문만 출력하기
re.findall(r'\([A-Za-z].+?[a-z|\.]\)',script101)[:6]
```

### 3)특정 단어의 예문만 모아 파일로 저장하기

```python
import os, re
os.chdir(r'C:\Users\user\Desktop\Plan A\.ipynb_checkpoints')
f = open('friends101.txt','r')
```

```python
#객체 f안의 모든 문장을 원소로 하는 리스트 만들기
sentences = f.readlines()
```

```python
#각 문장이 원소로 잘 들어갔는지 확인하기
sentences[:3]
```

```python
#대사만 추출해보기
for i in sentences[:20]:
    if re.match(r'[A-Z][a-z]+:',i):      #문장 맨 앞에서 패턴을 찾는 match문
        print(i)
```

```python
#대사만 담긴 리스트 만들기
lines = [i for i in sentences if re.match(r'[A-Z][a-z]+:',i)]
#sentences 객체의 원소 i중 if문을 충족하는 것을 그대로 lines 리스트에 저장
```

```python
lines[:10]
```

```python
#would가 들어간 문장 추출하기
would = [i for i in sentences if re.match(r'[A-Z][a-z]+:',i) and re.search('would',i)]
#if문 조건에 and re.search('would',i)를 추가해 would객체에 저장하기
```

```python
would
```

```python
#take가 들어간 문장 추출하기
take = [i for i in sentences if re.match(r'[A-Z][a-z]+:',i) and re.search('take',i)]
#if문 조건에 and re.search('take',i)를 추가해 take 객체에 저장하기
```

```python
take
```

```python
#take 객체의 원소를 차례대로 출력
for i in take:
    print(i)
```

```python
#추출한 문장을 파일에 저장
```

```python
newf = open('would.txt','w')      #'would.txt'파일을 쓰기모드(w)로 생성
newf.writelines(would)            # would 리스트의 원소를 꺼내 한 줄씩 쓰기
newf.close()                      # 파일 닫기
```
